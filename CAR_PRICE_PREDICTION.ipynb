```python
#percentage function
def without_hue(plot, feature):
    total = len(feature)
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
    plt.show()
```

### Data Preprocessing for model Training

#### Set car_ID  column as Index 

```python
df.set_index('car_ID', inplace=True)
```

#### Drop carname Column

```python
df = df.drop('CarName', axis='columns')
df.head()
```

```python
df.shape
```

### Convert categorical data into dummies and drop first option, since models dont't train on  categorical data

```python
df.columns
```

##### get all categorical columns from our dataset

```python
set_of_numeric = set(numeric_df.columns)
set_of_all = set(df.columns)
categorical_columns = set_of_all.symmetric_difference(set_of_numeric)
categorical_columns = list(categorical_columns)
print(categorical_columns)
```

```python
categorical_data = df[categorical_columns]
categorical_data.head()
```

```python
#use panda get_dummies method to convert categorical data into dummies
categorical_dummies = pd.get_dummies(categorical_data, drop_first=True)
```

```python
categorical_dummies.head()
```

```python
numeric_df.head()
```

```python
### concat dummies dataframe and numerci dataframe
full_df = pd.concat([numeric_df, categorical_dummies], axis = 1)

display(full_df.head())
print(full_df.shape)
```

#### Our final dataframe now has 44 columns

### Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','carlength','carwidth','price']
full_df[num_vars] = scaler.fit_transform(full_df[num_vars])
full_df.head()
```

## Build Linear regression model

```python
## import necessary libraries for model building
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
```

## Separate target varibales and feature variables

```python
X = full_df.drop('price', axis=1)
y = full_df['price']
```

#### split data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42,  shuffle=True)
```

#### Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print("Test R^2 Score  : ", model.score(X_test, y_test))
print("Train R^2 Score : ", model.score(X_train, y_train))
```

## Model Interpretation

```python
from sklearn.linear_model import Lasso
names = X_train.columns
lasso = Lasso(alpha=0.1)
lasso_coef = lasso.fit(X_train, y_train).coef_

plt.figure(figsize=(20,20))
ax = plt.plot(range(len(names)), lasso_coef)
ax = plt.xticks(range(len(names)), names, rotation=60)
ax = plt.ylabel('Coefficients')
plt.show()
```

```python
for i, (variable, coef) in enumerate(zip(names, lasso_coef)):
    print(i, variable, coef)
```
